## 1. 비즈니스 이해

### Background: 어떠한 문제가 존재하는지 상황에 대해 알려진 기사, 논문, 보고서 등을 인용하여 현상을 구체적으로 기술하시기 바랍니다.
- 위 기사에 따르면 투썸 플레이스는 2017년 커피 매장 고객 만족도 2위, 2018년 현재 1,000호점 개점을 앞둔 상태이다.
- '고객의 만족도가 2위인 투썸 플레이스가 과연 직원의 만족도도 고객 만족도만큼 높을까?' 에 대한 궁금증을 토대로 
학교 앞 가맹점인 '투썸 플레이스 꿈의 교회점' 직원 만족도와 고객 만족도를 설문조사를 통하여 구한 뒤 어떤 항목에서 차이가 벌어지는지 알아보고자 하였다. 

### Business Objectives: 기업(혹은 본인)에게 이 현상이 어떠한 이유로 분석을 해야 하는지 추상적(함축적)으로 기술하시기 바랍니다.
-  조원 중 한 명은 현재 투썸 플레이스 꿈의 교회점에서 1년이 넘는 기간 동안 아르바이트를 하고 있다. 가장 가까이서 손님을 응대하는 업무를 하다보면 실제로 1,000호점 개점을 앞둔 투썸 플레이스에 조차도 고객 응대 매뉴얼이 빈약하다는 생각이 들 때가 있다. 
그 예시로는 전문 바리스타 양성 교육을 들 수 있다. 바리스타 교육이 매 달 지역마다 실시되고 있지만, 한 교육 당 15만원이 드는 고가의 교육일 뿐더러 가맹점 매장 내 직원 중 신청자만 수강을 받을 수 있다. 따라서 나머지 직원 및 아르바이트생은 서면으로 전달되는 매뉴얼만 가질 뿐인데 처음 생성된 매뉴얼 외의 것들은 컴플레인을 받은 즉시 생성되는 매뉴얼이다. ‘밑 빠진 독에 물 붓기’라고 할 수 있는 매뉴얼들이다. 
투썸 플레이스 뿐 아니라, 스타벅스를 제외한 한국의 커피 체인점들 모두 비슷한 매뉴얼을 가지고 있는 것으로 알고 있다.
따라서 응대 매뉴얼(교육 훈련) 이수가 매장의 만족도에 영향을 미치는지에 대해 알아보고자 한다. 또한 기업 입장에서는 2015년 기준 투썸 플레이스의 가맹점 수는 633개였는데, 분석을 통해 가맹점에 이루어지는 현재의 본사 교육 빈도가 적정 수준인지 알아보려 한다.

### Business Success Criteria: 비즈니스 관점에서 분석결과의 성공기준(정량적) 또는 유용한 결과에 대한 기준(정성적)을 설명하십시오.

### Analysis Goals: 비즈니스 목표에 대한 의사결정 지원을 위해 실시하는 분석의 목표를 기술합니다. 
비즈니스 목표가 교육훈련 빈도가 적정수준인지 알아보기 위함을 바탕으로 분석의 목표는 교육훈련을 이수한 직원과 훈련을 이수하지 않은 직원을 분류
또한,  훈련 이수가 매장의 만족도에 영향을 미치는지에 대해서는 훈련 여부와 만족도와의 상관관계를 분석 및 훈련 여부에 따라 각 항목의 만족도 수치 비교

#### 2) 각 항목별로 교육 훈련을 받은 직원과 받지 않은 직원을 분류하여 만족도 비교

```python
# 교육훈련을 받은 직원과 받지 않은 직원으로 분류하여 인사 항목의 만족도를 비교
one = pd.pivot_table(staff, index=["교육 훈련"], values=["첫인사", "끝인사"])
one.head()
one_t = np.transpose(one)
one_t.head()
```

## 5. 시사점

상관관계 분석을 통해 분석 처음 매개 변수로 잡았던 교육훈련에 영향을 받는 항목은 '첫인사, 끝인사, 설명, 맛일정' 이 4가지 항목이었다.
설문지에서 받았던 근무기간도 매개변수2로 잡고 상관관계 분석을 한 결과로 '맛 일정, 토핑바 청결' 2가지 항목이 도출되었다.

따라서 두 매개의 관계에서 양의 상관관계를 가지는 '맛 일정' 항목이 직원의 만족도 항목에서 가장 높은 결과를 가짐을 알 수 있으며, 꼭짓점 그래프에서 확인할 수 있다.

우리는 
왜 고객들의 만족도보다 점원들의 만족도가 전반적으로 더 낮은가? 와 상관관계와 관련 없는 항목들은 무슨 변수 때문에 만족도가 낮은가? 에 대해 알아봐야 한다.

상대적으로 다른 항목들에 비해 인사와 관련된 항목(첫인사, 끝인사)는 고객과 직원 간의 차이가 크지 않았다. 다만 첫인사보다 끝인사의 고객 만족도가 더 높았으나, 점원들은 두 항목 모두에 같은 결과값을 가지고 있다. 이는 점원은 같은 정도로 인사를 했다고 인식하지만 실제로 고객은 퇴장 인사를 입장 인사보다 더 많이 들었을 수 있다. 따라서 입장하는 순간에 느껴지는 매장의 만족도에 개선이 필요하다는 것을 알 수 있다.

다음은 말투이다. 
말투의 설문은 [주문 시 느끼는 점원의 말투 만족도]인데, 손님과 점원의 차이가 가장 크게 나는 항목이다. 손님은 주문 시 점원의 말투(어투)가 다른 항목들에 비해서도 가장 만족스러운데 (4.28점), 점원은 3.56의 평균 정도의 만족도를 가진다.
더군다나 말투는 상관관계 분석 시 교육 훈련과 근무 기간 중 어느 하나에도 영향을 받지 않았다. 따라서 다른 변수가 말투에 영향을 미친다는 것인데, 이는 그간 근무를 하며 점주의 영향을 받았다고 판단된다.
실제로 말투에 대해 손님으로부터 컴플레인이 없었던 한 직원에게 ‘ㅇㅇ아. 주문 받을 때 항상 말투 조심해.’라는 말을 들은 직원이 있다. 다른 직원에게는 ‘도레미파 솔! 톤’으로 주문을 받으라는 지적을 받았다고 한다. 조원에게는 ‘어미를 왜 이렇게 올리냐.’ , ‘목소리 톤이 너무 낮은 거 아니냐.’ 라는 등의 지적을 받은 적이 있었다. 
지금은 퇴사한 예전 매니저는 와이파이 비밀번호를 물어보는 손님께 조원이 ‘숫자 1부터 8까지 입력하시면 돼요.’ 라고 한 것을 보며 ‘손님은 그렇게 말하면 못 알아듣는다. 1,2,3,4,5,6,7,8이라고 말해라.’ 라는 지적을 받은 적도 있다. 
여기서 드는 의문점은 ‘화장실은 우측으로 가시면 나오세요.’라든가, ‘고객님 음료 나오셨습니다.’와 같은 국어사전에 맞지 않는 존대에 대해서는 어떠한 지적도 받지 않았다는 것이다. 잘못된 말에 올바른 지적이 있었다면 점원의 말투 만족도가 이 정도로 낮지는 않았을 것이라 판단된다.
다음은 성실성에 대한 분석이다. 
조원이 처음 아르바이트를 시작하던 날 매니저에게 들었던 말은 ‘점주님은 가만히 있는 걸 싫어하세요.’였다. 그 뒤로 다른 점원이 일을 시작했을 때도 빠지지 않고 전달된 말이다. 따라서 할 일을 끝내고 가만히 있으면 항상 지적을 받아왔고, 따라서 만족도가 낮을 수밖에 없을 거라고 판단된다.
이는 설문을 시작하기 전부터 조원들이 예상했던 부분이기도 하다. 
반면에 손님들의 성실성 만족도는 다른 항목과 비교해서도 굉장히 높은 편인데, 이는 손님 의견란에 기재되어 있었던 ‘음료가 빠르게 나와서 좋다.’ , ‘다른 매장에 비해 음료가 빠르게 나온다.’등으로 설명할 수 있다. 
손님의 주문을 도와주는 점원이 결제를 진행하는 동안 다른 점원들은 쇼케이스에서 케이크를 꺼내고 커피를 내리며 음료를 만든다. 그래서 만드는데 복잡하지 않은 음료 같은 경우엔 결제가 끝나기도 전에 음료를 완성할 수도 있다. 
이 부분에서 아쉬운 것은 매장이 한창 바쁠 시간대에는 아무도 가만히 있지 않고 각자 자신의 할 몫을 다 함에도 불구하고 성실성에 대한 계속된 지적이 이어져 손님에게는 올바른 서비스를 제공하고도 점원의 만족도가 낮다는 점이다.

다음은 설명에 대한 시사이다
교육훈련과 상관관계가 있는 항목이기도 한 설명 부분은 상대적으로 교육 훈련을 받은 점원은 손님에게 잘 응대한다고 인식하는 반면에 교육 훈련을 받지 않은 점원은 그렇지 않은 인식 수준을 가지고 있다. 이는 교육 훈련을 통해 모든 점원의 인식이 높아질 필요가 있다고 판단된다.

맛 일정에 관한 시사.
맛 일정은 유일하게 손님보다 점원의 만족도가 더 높다. 이는 근무 기간과 교육 훈련 두 가지의 매개 변수에 모두 상관관계 영향을 받기 때문이라고 설명할 수 있겠다.
또한 실제로도 어떤 재료를 넣으면 음료의 당도가 낮아지는지, 높아지는지, 어떤 재료가 해당 음료의 풍미를 더 살리는지 등에 대해서 교육을 통해 이수 받기도 하지만 장기 근무를 하며 자연스레 터득할 수 있는 부분이기도 하다. 따라서 장기 근무자가 많은 해당 매장에서는 맛 일정 부분에 대한 점원의 만족도가 높을 수밖에 없다고 판단된다. 

비품과 토핑바 청결은 시설물 부문으로 함께 묶을 수 있다.
해당 매장은 교회 앞에 위치했다는 특성 상 일요일 교회 예배가 마치는 시간이 일주일 중 가장 바쁜 시간이다. 시간당 가장 많은 점원이 일하지만 가장 일손이 부족한 시간대이다. 점주는 준비된 음료를 손님에게 전달한다. 그 뒤에서는 한 명은 주문을 받고, 한 명은 커피를 내리며, 한 명은 케이크를 준비하고 음료를 만든다. 그러면 휴지나 물티슈, 물 등이 모자랄 때가 있고 손님들이 토핑바에 컵을 그냥 둔 채로 나가게 되면 그것을 채우고 치울 일손이 부족하다. 점주는 당장에 보이는 문제에 대해 토핑바 정리 필요하다, 물 채워야 한다. 등의 주문을 넣지만, 각자에게 정해진 일을 하느라 비품을 제 시간에 채워넣지 못한다거나, 토핑바 위를 제 때에 치울 수 없는 경우가 빈번하다.
또한 하루동안 모아진 쓰레기는 그 날 밤에 점원 중 한 명이 분리수거를 한다. 플라스틱, 캔, 유리, 종이를 모두 분리하게 되는데, 쓰레기를 뒤지는 업무이다보니 아무리 근무 환경이 쾌적하다고 하더라도 토핑바에 대한 만족도는 현저히 떨어질 수밖에 없다. 

## 총 결론

현재 카페에서 근무하고 있는 경험을 바탕으로 고객의 만족도 보다 직원의 만족도가 낮은 이유로 조직문화를 꼽았다. 조사한 논문에서는 관계문화를 가진 기업일수록 구성원들의 일 내용 만족도 수준, 근속몰입 수준, 감정몰입 수준이 높다고 밝히고 있었다. 꼭 조사한 논문이 아니더라도 올바른 조직문화의 형성은 기업의 지속가능성을 창출할 수 있는 밑거름이라는 것을 익히 알고 있는 바이다. 따라서 설문결과에서 보여진 직원의 낮은 만족도를 개선하기 위해 직원들의 자긍심 고취와 긍정적인 조직 분위기를 형성하여 직원들이 보다 본업에 몰입할 수 있는 환경이 필요하다.
